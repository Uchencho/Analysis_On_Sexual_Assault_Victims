From the result above, looks like the data was collated in April 2019, I will create a day and time column to see if there is any interesting pattern there.

```python
import datetime as dt

data['day'] = data['time'].dt.day                                #extracted the day
data["hour"] = data["time"].dt.hour                              #extracted the hour

data.head()
```

I have created two new columns, the day and hour coulmn which was extracted from the time column

```python
data['hour'].value_counts(dropna = False)
```

From the hour column, I will create a new column that allows us to see easily if the data was collated in the early hours of the day, morning, afternoon, evening or night

```python
def hour_segment(x):
    if 0 <= x < 6:
        return "early_morning"
    elif 6 <= x < 12:
        return "morning"
    elif 12 <= x < 16:
        return "afternoon"
    elif 16 <= x < 19:
        return "evening"
    elif 19 <= x < 24:
        return "night"

data['period_of_day'] = data['hour'].apply(hour_segment)
data['period_of_day'].value_counts(dropna = False)
```

We have created extra columns highlighting the day of April 2019, hour(24-hr format) and period of the day that the data was collated. The minute and second was not extracted because it was not deemed necessary. We will continue cleaning the data by looking at more columns, next is the age column.
