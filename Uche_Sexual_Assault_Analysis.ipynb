__From the result above, looks like the data was collated in April 2019, I will create a day and time column to see if there is any interesting pattern there.__

```python
import datetime as dt

data['day'] = data['time'].dt.day                                #extracted the day
data["hour"] = data["time"].dt.hour                              #extracted the hour(24-hr format)

data.head()
```

__I have created two new columns, the day and hour coulmn which was extracted from the time column__

```python
data['hour'].value_counts(dropna = False)
```

__From the hour column, I will create a new column that allows us to see easily if the data was collated in the early hours of the day, morning, afternoon, evening or night__

```python
def hour_segment(x):
    if 0 <= x < 6:
        return "early_morning"
    elif 6 <= x < 12:
        return "morning"
    elif 12 <= x < 16:
        return "afternoon"
    elif 16 <= x < 19:
        return "evening"
    elif 19 <= x < 24:
        return "night"

data['period_of_day'] = data['hour'].apply(hour_segment)
data['period_of_day'].value_counts(dropna = False)
```

__We have created extra columns highlighting the day of April 2019, hour(24-hr format) and period of the day that the data was collated. The minute and second was not extracted because it was not deemed necessary. We will continue cleaning the data by looking at more columns, next is the gender column.__

__Although tempted to convert this into a binary response, the second result with 320 occurrences will be difficult to convert to binary because the response is not definite, hence we will leave it.__

```python
data['commercial_dna_services'].value_counts(dropna = False)
```

__This is great, i will convert this column into Boolean expression__

```python
yes_no = {'Yes' : True, 'No' : False}
data['commercial_dna_services'] = data['commercial_dna_services'].map(yes_no)
data['commercial_dna_services'].value_counts(dropna = False)
```

```python
data['witness_assault/rape'].value_counts(dropna = False)
```

__Same thing, i will convert this too boolean too and set the NaN value to false__

```python
yes_no_NaN = {'Yes' : True, 'No' : False, np.NaN: False}
data['witness_assault/rape'] = data['witness_assault/rape'].map(yes_no_NaN)
data['witness_assault/rape'].value_counts(dropna = False)
```
